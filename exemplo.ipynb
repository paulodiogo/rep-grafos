```python
from rep_grafos import RepModular, RepProbabilistic, RepSpatialXOR
import igraph
```

```python
N = 100
d = 0.5
M = int(((N*(N-1))/2)*d)
G = igraph.Graph.Erdos_Renyi(n=N, m=M)
```

```python
c = 3
size = 16
q = 0.12

rep_1 = RepProbabilistic(c, size, q)    
rep_1.construir(G)

rep_1.checks()
```

```python
c = 3

rep_2 = RepModular(c, minimo=100)    
rep_2.construir(G)

rep_2.checks()
```

```python
hash_size = 8
k = 5

rep_3 = RepSpatialXOR(hash_size, k)    
rep_3.construir(G)

rep_3.checks()
```
